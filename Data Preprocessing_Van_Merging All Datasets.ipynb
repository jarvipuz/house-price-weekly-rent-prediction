```python
import pandas as pd
import numpy as np
```

# Merging Rent & Sales 2019

```python
#paths
rent19_bedroom_path = "Data/house_rent_sales/house_rent_0619_bedroom.csv"
rent19_dwelling_path = "Data/house_rent_sales/house_rent_0619_dwelling.csv"
rent19_overall_path = "Data/house_rent_sales/house_rent_0619_overall.csv"

sales19_dwelling_mean_path = "Data/house_rent_sales/house_sales_0319_dwelling_mean.csv"
sales19_dwelling_median_path = "Data/house_rent_sales/house_sales_0319_dwelling_median.csv"
sales19_overal_mean_path = "Data/house_rent_sales/house_sales_0319_overall_mean.csv"
sales19_overal_median_path = "Data/house_rent_sales/house_sales_0319_overall_median.csv"
```

```python
#Load and merge data
rent19_bedroom = pd.read_csv(rent19_bedroom_path)
rent19_dwelling = pd.read_csv(rent19_dwelling_path)
rent19_overall = pd.read_csv(rent19_overall_path)

sales19_dwelling_mean = pd.read_csv(sales19_dwelling_mean_path)
sales19_dwelling_median = pd.read_csv(sales19_dwelling_median_path)
sales19_overal_mean = pd.read_csv(sales19_overal_mean_path)
sales19_overal_median = pd.read_csv(sales19_overal_median_path)
```

```python
rent19 = rent19_bedroom.merge(rent19_dwelling, on=['LGA_CODE', 'LGA_NAME'], how='outer')
rent19 = rent19.merge(rent19_overall, on=['LGA_CODE', 'LGA_NAME'], how='outer')

sales19 = sales19_dwelling_mean.merge(sales19_dwelling_median, on=['LGA_CODE', 'LGA_NAME'], how='left')
sales19 = sales19.merge(sales19_overal_mean, on=['LGA_CODE', 'LGA_NAME'], how='left')
sales19 = sales19.merge(sales19_overal_median, on=['LGA_CODE', 'LGA_NAME'], how='left')
```

```python
rent19.head()
```

```python
rent19.columns
```

```python
rent19.columns = ['LGA_CODE', 'LGA_NAME', 'r_one_br', 'r_two_br', 'r_three_br',
                  'r_four_or_more_br', 'r_bedsitter', 'r_not_specified', 'r_flat', 'r_house',
                  'r_other', 'r_townhouse', 'r_weekly_rent_median']
```

```python
rent19.head()
```

```python
sales19.columns
```

```python
sales19.columns = ['LGA_CODE', 'LGA_NAME', 's_mean_nonstrata', 's_mean_strata',
                   's_median_nonstrata', 's_median_strata', 's_mean_sales', 's_median_sales']
```

```python
sales19.head()
```

```python
rent_sales19 = sales19.merge(rent19, on=['LGA_CODE', 'LGA_NAME'], how='outer')
```

```python
rent_sales19.head()
```

```python
rent19.to_csv("Data/house_rent_sales/final/rent19_final.csv", index = None, header=True)
sales19.to_csv("Data/house_rent_sales/final/sales19_final.csv", index = None, header=True)
rent_sales19.to_csv("Data/house_rent_sales/final/rent19_sales19_final.csv", index = None, header=True)
```

# Merging Rent & Sales 2017

```python
#paths
rent17_bedroom_path = "Data/house_rent_sales/house_rent_0917_bedroom.csv"
rent17_dwelling_path = "Data/house_rent_sales/house_rent_0917_dwelling.csv"
rent17_overall_path = "Data/house_rent_sales/house_rent_0917_overall.csv"

sales17_dwelling_mean_path = "Data/house_rent_sales/house_sales_0617_dwelling_mean.csv"
sales17_dwelling_median_path = "Data/house_rent_sales/house_sales_0617_dwelling_median.csv"
sales17_overal_mean_path = "Data/house_rent_sales/house_sales_0617_overall_mean.csv"
sales17_overal_median_path = "Data/house_rent_sales/house_sales_0617_overall_median.csv"
```
